```python
from duckduckgo_search import ddg_images
from fastcore.all import *
```

```python
def search_images(term, max_images=150):
    return L(ddg_images(term, max_results=max_images)).itemgot('image')
```

```python
urls=search_images('grizzly bear', max_images=1)
urls[0]
```

```python
from fastdownload import download_url
dest='grizzlybear.jpg'
download_url(urls[0], dest, show_progress=False)
```

```python
from fastai.vision.all import *
im=Image.open(dest)
im.to_thumb(256,256)
```

```python
download_url(search_images('teddy bear', max_images=1)[0], 'teddybear.jpg', show_progress=False)
Image.open('teddybear.jpg').to_thumb(256,256)
```

```python
download_url(search_images('black bear', max_images=1)[0], 'blackbear.jpg', show_progress=False)
Image.open('blackbear.jpg').to_thumb(256,256)
```

```python
bear_types='black bear', 'teddy bear', 'grizzly bear'
path=Path('bears')
```

```python
for o in bear_types:
    dest = (path/o)
    dest.mkdir(exist_ok=True, parents=True)
    results = search_images(f'{o} photos')
    download_images(dest, urls=results[:150])
    resize_images(path/o, max_images=450, dest=path/o)
        
```

```python
fns = get_image_files(path)
fns
```

```python
failed = verify_images(fns)
len(failed)
```

```python
failed.map(Path.unlink)
```

```python
bears=DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=Resize(128))
```

```python
dls=bears.dataloaders(path)
```

```python
dls.show_batch(max_n=4, nrows=1)
```

```python
bears=bears.new(item_tfms=Resize(128, ResizeMethod.Squish))
dls=bears.dataloaders(path)
dls.show_batch(max_n=4, nrows=1)
```

```python
bears=bears.new(item_tfms=RandomResizedCrop(128, min_scale=0.3))
dls=bears.dataloaders(path)
dls.train.show_batch(max_n=4, nrows=1, unique=True)
```

## Data Augmentation

```python
bears=bears.new(item_tfms=Resize(128), batch_tfms=aug_transforms(mult=2))
dls=bears.dataloaders(path)
dls.train.show_batch(max_n=8, nrows=2, unique=True)
                
```

## Training Model

```python
bears=bears.new(
    item_tfms=RandomResizedCrop(224, min_scale=0.5),
    batch_tfms=aug_transforms())
dls=bears.dataloaders(path)
```

```python
learn = vision_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(4)
```

```python
interp=ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(5, nrows=2, figsize=(12,10))
```

## Data Cleaning

```python
from fastai.vision.widgets import *
cleaner=ImageClassifierCleaner(learn)
cleaner
```

```python
for idx in cleaner.delete(): cleaner.fns[idx].unlink()
for idx,cat in cleaner.change(): shutil.move(str(cleaner.fns[idx]), path/cat)
```

## Export the model

```python
learn.export()
```

```python
path = Path()
path.ls(file_exts='.pkl')
```

## creating an app from the notebook

```python
btn_run = widgets.Button(description='Classify')
btn_run
```

```python
def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(256,256))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)
```
